```python
somaxvalues=pd.read_csv('/content/drive/MyDrive/Madonna/data_log57.csv',index_col=[0])
somaxvalues
somadata=pd.read_csv('/content/drive/MyDrive/Madonna/SomaticPointMutationMatrix.csv',index_col=[0])
somadata
copydata=pd.read_csv('/content/drive/MyDrive/Madonna/CopyNumberAlterationMatrix.csv',index_col=[0]) # here we make first col of data as index col of the data
copydata
copyxvalues=pd.read_csv('/content/drive/MyDrive/Madonna/data_copy.csv',index_col=[0])
copyxvalues

print(somaxvalues)
print(somadata)
print(copydata)
print(copyxvalues)
```

# 1.Selected Features Somatic Muatation Matrix 








```python
extract_soma= somadata[somaxvalues.values.ravel()]
extract_soma
```

# 2.Selected Features Copy Number Alteration Matrix

```python
copyxvalues.values.ravel()
```

```python
extract_copy= copydata[copyxvalues.values.ravel()]
extract_copy
```

```python
extract_copy
```

# 3.Combining Somatic mutation matrix and Copy number alteration matrix

```python
df = pd.concat([extract_soma,extract_copy], axis=1)
df
```

```python
target = 'classLabels'
X = df
Y = somadata.loc[:, somadata.columns == target]
Y = Y.values.ravel()
print(X)
print(Y)
```

# 4.Splitting of dataset

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.33, 
                                                    random_state=8)
```

# 5.Applying logistic regression

```python
clf_lr = LogisticRegression(max_iter=5000)      
lr_baseline_model = clf_lr.fit(x_train,y_train)
```

```python
ac = accuracy_score(y_test,lr_baseline_model.predict(x_test))
print(ac)
```

# 6.Applying recursive feature selection

```python
rfecv = RFECV(estimator=clf_lr, step=100, cv=3, scoring='accuracy')
rfecv = rfecv.fit(x_train, y_train)
print('Optimal number of features :', rfecv.n_features_)
print('Best features :', x_train.columns[rfecv.support_])
```

```python
x_train_rfecv = rfecv.transform(x_train)
x_test_rfecv = rfecv.transform(x_test)
```

```python
lr_rfecv_model = clf_lr.fit(x_train_rfecv, y_train)
```

```python
ac = accuracy_score(y_test,lr_rfecv_model.predict(x_test_rfecv))
ac
```
